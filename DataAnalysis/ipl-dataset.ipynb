```python
df1=ipl_data[["PLAYER NAME","SOLD PRICE"]]
print(df1)
df1.sort_values(axis=0,by="SOLD PRICE",inplace=True)
print(df1)
```

```python
ipl_data["PREMIUM"]=ipl_data["SOLD PRICE"]-ipl_data["BASE PRICE"]
ipl_data
```

```python
max_premium_players=ipl_data[["PLAYER NAME","COUNTRY","PREMIUM"]]
max_premium_players.sort_values("PREMIUM",inplace=True,ascending=False)
max_premium_players
```

```python
average_sold_per_age=ipl_data.groupby("AGE")["SOLD PRICE"].mean()
average_sold_per_age=average_sold_per_age.reset_index()
```

```python
average_sp_per_age_playingrole=ipl_data.groupby(["AGE","PLAYING ROLE"])["SOLD PRICE"].mean()
average_sp_per_age_playingrole=average_sp_per_age_playingrole.reset_index()
```

```python
average_sold_per_age
```

```python
average_sp_per_age_playingrole
```

```python
soldprice_comparison=pd.merge(left=average_sold_per_age,right=average_sp_per_age_playingrole,on="AGE",how="outer")
soldprice_comparison
```

```python
soldprice_comparison.rename(columns={"SOLD PRICE_x":"SOLD PRICE FOR AGE","SOLD PRICE_y":"SOLD PRICE FOR PLAYING ROLE"},inplace=True)
soldprice_comparison
```

```python
soldprice_comparison["CHANGE"]=soldprice_comparison.apply(lambda x: (x["SOLD PRICE FOR AGE"]-x["SOLD PRICE FOR PLAYING ROLE"])/x["SOLD PRICE FOR AGE"],axis=1)
soldprice_comparison
```
