```python
### Data visualization using seaborn
# seaborn is a python visualization library based on matplotlib
# it helps in creating complex visualizations, in just few lines of code.
```

```python
import seaborn as sns
```

```python
# basic plotting
# loading a default dataset from seaborn
tips_data=sns.load_dataset("tips")
tips_data # tips is a default dataset present in seaborn github
```

```python
# creating a scatter plot of 'tip' vs 'total bill' 
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',marker='o',data=tips_data) # plotting the scatterplot for column->'total_bill' and 'tip' and passing data as->tips_data, meaning extract the columns specified from this data
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs tip')
plt.show()
```

```python
# creating a line plot
# for total_bill vs size(no of people)
sns.lineplot(x='size',y='total_bill',data=tips_data)
plt.xlabel('No of People')
plt.ylabel('Total Bill')
plt.show()
```

```python
## Categorical Plots-> the plots which are plotted for categorical values
# columns having no numeric values eg-> sex,smoker,day,time
```

```python
# Bar plot of 'total_bil' vs 'day'
sns.barplot(x='day',y='total_bill',data=tips_data) # same format
plt.title("Total Bill vs Day")
plt.show()
```

```python
# Next plot is Box Plot
# box plot is created with five numbers-> min,first quartile,median,third quartile,max
```

```python
# box plot of 'day' vs 'total_bill'
sns.boxplot(x="day",y="total_bill",data=tips_data)
plt.title("Day vs Total Bill")
plt.show()
```

```python
# we can also get a violin plot
# the violin plot, will give an estimate of the distribution of data, with respect to the total_bill
sns.violinplot(x="day",y="total_bill",data=tips_data)
plt.show()
```

```python
# creating a histogram for the total_bill
sns.histplot(x="total_bill",data=tips_data,bins=10,kde=True) # kde='true' shows you the density estimation curve for the histogram
```

```python
# if you want to see only the KDE plot, apart from histogram
sns.kdeplot(x="total_bill",data=tips_data,fill=True)
plt.show()
```

```python
# pair plot
# using pair plot you can see the relationship between every other feature present in the dataset
# in a single row -> common y-axes
# in a single column-> common x-axes
sns.pairplot(tips_data)
plt.show()
# for the pair plot, no categorical variable is taken into consideration
```

```python
# next we will find out correlation between certain numerical features
# correlation can be applied only between numerical features
```

```python
corr_data=tips_data[["total_bill","tip","size"]].corr() # we'll find out the relation between these three features
corr_data
```

```python
# here see that the correlations
# total_bill vs tip -> 0.675734 , which means they are moderately positive, hence as total bill increases the tip tends to increase
# total_bill vs size -> 0.598315, which means moderately positive, since bigger groups tend to have bigger bills
# tip vs size -> 0.489299, which means moderate correlation, bigger groups tend to leave bigger bills, but not so strong relationship
```

```python
# now based on this correlation, we'll see the heatmap
sns.heatmap(corr_data,cmap="coolwarm")
# see that the color and heatmap is being displayed, but we are not able to see the correlation value
```

```python
# to see the correlation value, set annot=true
sns.heatmap(corr_data,cmap="coolwarm",annot=True)
plt.show()
# now the we can see the numbers
# in this color map: red-> highest correlation
                  #  and then as it moves to darkest shade of blue-> the correlation decreases
```

```python
import pandas as pd
sales_data=pd.read_csv("sales_data.csv")
sales_data.head(5)
```

```python
# now i will plot "total revenue" by "product category"
plt.figure(figsize=(10,10))
sns.barplot(x="Product Category",y="Total Revenue",data=sales_data,estimator=sum) 
# here by default, the estimator was set to mean
# meaning, for each category, the mean of total revenue was calculated, and plotted accordingly
# i have set it to sum, so that now, the sum of total revenue for each category is calculated, and plotted accordingly
plt.show()
```

```python
# getting sum of total revenues by region
plt.figure(figsize=(5,5))
sns.barplot(x="Region",y="Total Revenue",data=sales_data,estimator=sum)
plt.show()
```
