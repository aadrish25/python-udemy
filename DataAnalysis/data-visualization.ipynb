```python
### Data Visualization using matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
# create a line plot with x and y axis
# define the x axis
x=[1,2,3,4,5]
# define the y axis
y=[1,4,9,16,25]
```

```python
# now create the line plot
plt.plot(x,y) # this will create the line plot with x and y axis
plt.xlabel('X Axis') # this will provide a label to x-axis
plt.ylabel('Y Axis') # this will provide a label to y-axis
plt.title('Basic Line PLot') # this will provide a title to the plot
plt.show() # finally this will show the plot
```

```python
# customizing your line plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# create a red colored line plot 
plt.plot(x,y,color="red")
```

```python
# create a double dashed line plot '--'
plt.plot(x,y,color="red",linestyle="--")
# see that the line style is '--' as we have mentioned
```

```python
# now see the marker
plt.plot(x,y,color="red",linestyle='--',marker='o')
# it means that at every (x,y) co-ordinate according to our provided data, there will be a 'o' mark
```

```python
# you can also control the width of the line as well as the marker size
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=10)
# the line will appear much thicker as well as the marker will appear much bigger
# also if we set grid(True) it will make a grid in our plot
plt.grid(True)
```

```python
## Creating multiple plots using subplot
# defining data for multiple plot
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,8,27,64,125]
```

```python
# now in order to fit multiple plots, we will define a parent figure
plt.figure(figsize=(9,5))
# we create a parent figure with 900x500 pixels 
# this figure will contain the subplots
```

```python
# defining the first subplot
# the position of the subplot inside the figure is defined in this format->subplot(no of rows,no of columns,position of the current subplot)
plt.subplot(1,2,1) # this means the figure will have 1->row, 2->columns and position of this subplot is->1
plt.plot(x,y1,color="green")
plt.title("Plot 1") 


# now let's define the position of next subplot
plt.subplot(1,2,2) # this means the figure will have 1->row,2->columns and position of this subplot is ->2
plt.plot(x,y2,color="red")
plt.title("Plot 2")

# see that the two subplots are in their corresponding positions, and there are exactly 1 row and 2 columns in the parent figure
```

```python
# creating 4 subplots
x=[1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[1,4,9,16,25]
y3=[1,8,27,64,125]
y4=[10,20,30,40,50]
# defining the axes for 4 subplots

# making the parent figure
plt.figure(figsize=(11,10))
```

```python
# creating the first subplot
# there will be 2 rows and 2 columns, so the format will be like->(2,2,position of current subplot)
plt.subplot(2,2,1)
plt.plot(x,y1,color="green")
plt.title("Plot 1")

# creating the second subplot
plt.subplot(2,2,2)
plt.plot(x,y2,color="red")
plt.title("Plot 2")

# creating the third subplot
plt.subplot(2,2,3)
plt.plot(x,y3,color="blue")
plt.title("Plot 3")

# creating the fourth subplot
plt.subplot(2,2,4)
plt.plot(x,y4,color="orange")
plt.title("Plot 4")
```

```python
### Bar Plot
# suppose we have some categories, and with respect to those categories,I want to map some values
categories=['A','B','C','D','E']
values=[5,6,7,3,8]
# i want to map the values such that, A->5
                                    # B->6
                                    # C->7
                                    # D->3
                                    # E->8
# for this purpose I will create a bar plot
```

```python
# creating a bar plot
plt.bar(categories,values,color="purple")
# see that how beautifully,a bar plot is created
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

```python
### Histograms
# used to represent the distribution of a dataset
# the data is divided into bins, and the no of data points in each bin is counted
```

```python
# lets understand by an example
# below given are the exam scores of students
exam_scores=[45,48,52,56,59,62,66,70,72,75]
# now if we divide them into 4 bins, what will be the distribution?
# 40-49->2 students (bin 1)
# 50-59->3 students (bin 2)
# 60-69->2 students (bin 3)
# 70-79->3 students (bin 4) 
```

```python
# so the histogram plot will have 4 bars, and their height will be->2,3,2,3
# lets plot the histogram
plt.hist(exam_scores,bins=[40,50,60,70,80]) 
# see that i have passed a list-> declaring the interval of bins
# 40-to-50=>excluding 50
# 50-to-60=>excluding 60
# 60-to-70=>excluding 70
# 70-to-80=>excluding 80
```

```python
# and to view the diagram a little better and distinguished
plt.hist(exam_scores,bins=[40,50,60,70,80],edgecolor="black",color="red")
```

```python
### Pie Charts
# used to visulaize how much a particular category, occupies out of 100
categories=['A','B','C','D']
# so there are 4 categories
sizes=[30,20,40,10]
# A->30% out of 100%
# B->20% out of 100%
# C->40% out of 100%
# D->10% out of 100%
# we will visualize this with the help of a pie chart
colors=["orange","yellow","lightblue","pink"] # these are colors to distinguish different parts of the pie chart 
```

```python
# creating the pie char
plt.pie(sizes,labels=categories,colors=colors)
plt.show()
```

```python
# we can also show the precentage occupied by each category, and add shadow effect
plt.pie(sizes,labels=categories,colors=colors,autopct="%.1f%%",shadow=True)
# autopct format-> '%'-> marks the start of format specifier
                #  '.1'-> signifies that there will be only one place after the decimal point
                #  '%'->it is given before the end '%' so that the '%' sign is actually printed along with the nos.
                #  '%' -> it specifies the end of format specifier 
plt.show()
```

```python
# also we add explode property to make one or more categories stand out from the pie chart
explode_parameter=(0.2,0,0,0) # since only the first one is set, so only first category will pop outo f pie chart
plt.pie(sizes,labels=categories,explode=explode_parameter,colors=colors,autopct="%.1f%%",shadow=True)
# see that how the category 'A' has popped out of the pie chart
plt.show()
```

```python
### Now practical example on sales data visualization
import pandas as pd
sales_data_df=pd.read_csv("sales_data.csv")
sales_data_df.head(5)
```

```python
# we'll visualize, the total units sold, based on product category
product_categorywise_units=sales_data_df.groupby('Product Category')['Units Sold'].sum()
print(product_categorywise_units)
```

```python
# plotting it onto a bar plot
product_categorywise_units.plot(kind="bar",color="cyan")
plt.show()
```

```python
# plot sales trend over time
sales_over_time=sales_data_df.groupby("Date")["Total Revenue"].sum().reset_index() # we do reset index so that we have two columns
print(sales_over_time)
# see now that we have two columns -> 'Date' and 'Total Revenue'
```

```python
# we can plot like this
plt.plot(sales_over_time['Date'],sales_over_time['Total Revenue'],color="red")
plt.show()
```
