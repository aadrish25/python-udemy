```python
sns.countplot(data=netflix_data,x="type")
plt.xlabel("Movies and TV Shows")
plt.show()
```

```python
sns.histplot(data=netflix_data,x="release_year",bins=20)
plt.show()
```

```python
countries_and_titles=netflix_data[["title","country"]]
top_10_countries=countries_and_titles.value_counts("country").head(10).reset_index()
```

```python
top_10_countries
```

```python
plt.figure(figsize=(10,11))
sns.barplot(data=top_10_countries,x="country",y="count")
plt.show()
```
