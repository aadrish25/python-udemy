```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
worldcup_data=pd.read_csv("WorldCupMatches.csv")
```

```python
worldcup_data.head(5)
```

```python
worldcup_data.info()
print(f"Total Matches: {worldcup_data.shape[0]}")
```

```python
print(f"Earliest Year: {worldcup_data["Year"].min()}")
print(f"Latest Year : {worldcup_data["Year"].max()}")
```

```python
worldcup_data.value_counts("Stadium")
```

```python
worldcup_data.value_counts("City")
```

```python
worldcup_data["Total Goals"]=worldcup_data["Away Team Goals"]+worldcup_data["Home Team Goals"]
```

```python
worldcup_data.head(5)
```

```python
average_goals=worldcup_data["Total Goals"].mean()
print(f"Average Goals per match: {int(average_goals)}")
```

```python
matches_with_total_goals=worldcup_data[["Home Team Name","Away Team Name","Total Goals"]]
matches_with_total_goals
```

```python
matches_with_total_goals.sort_values(axis=0,by="Total Goals",ascending=False).head(5)
```

```python
worldcup_data.info()
```

```python
average_attendance=worldcup_data["Attendance"].mean()
print(f"Average attendance per match: {int(average_attendance)}")
```

```python
worldcup_data["Goal Difference"]=abs(worldcup_data["Home Team Goals"]-worldcup_data["Away Team Goals"])
```

```python
worldcup_data.head(5)
```

```python
home_goals=worldcup_data.groupby("Home Team Name")["Home Team Goals"].sum()
away_goals=worldcup_data.groupby("Away Team Name")["Away Team Goals"].sum()
```

```python
home_goals
```

```python
away_goals
```

```python
total_goals=home_goals.add(away_goals)
total_goals
```

```python
top_ten=total_goals.sort_values(axis=0,ascending=False).head(10)
top_ten
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=top_ten)
plt.show()
```

```python
worldcup_data
```

```python
sns.histplot(worldcup_data["Total Goals"])
plt.show()
```

```python
total_goals_per_tournament=worldcup_data.groupby("Year")["Total Goals"].sum().reset_index(name="Total Goals")
total_goals_per_tournament
```

```python
average_goals_per_tournament=worldcup_data.groupby("Year")["Total Goals"].mean().reset_index(name="Average Goals").astype(dtype=int)
average_goals_per_tournament
```

```python
sns.lineplot(data=average_goals_per_tournament,x="Year",y="Average Goals")
plt.xlabel("Years")
plt.ylabel("Average Goals")
plt.show()
```
