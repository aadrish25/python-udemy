```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_df=pd.read_csv("spotify_tracks.csv")
spotify_df
```

```python
unique_artist_tracks=spotify_df.value_counts(["artists","track_id"])
unique_artist_tracks
```

```python
spotify_df.value_counts("artists").sort_values(ascending=False)
```

```python
spotify_df["duration_min"]=spotify_df["duration_ms"]/(1000*60)
spotify_df
```

```python
plt.figure(figsize=(10,10))
sns.histplot(data=spotify_df,x="duration_min",bins=100,binrange=(0,20))
plt.show()
```

```python
spotify_df
```

```python
print(f"Avg popularity: {spotify_df["popularity"].mean()}")
```

```python
plt.figure(figsize=(10,10))
sns.histplot(data=spotify_df,x="popularity")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=spotify_df.head(1000),x="danceability",y="energy")
plt.show()
```

```python
most_popular_ten=spotify_df[["popularity","track_name","artists","tempo"]].sort_values(by="popularity",ascending=False).head(10)
most_popular_ten
```

```python
print(f"Average temp of top ten: {most_popular_ten["tempo"].mean()}")
```

```python
sns.histplot(spotify_df["valence"])
plt.show()
```

```python
highest_loudness=spotify_df.loc[spotify_df["loudness"].idxmax(),["track_name"]]
highest_loudness
```

```python
lowest_loudness=spotify_df.loc[spotify_df["loudness"].idxmin(),["track_name"]]
lowest_loudness
```

```python
spotify_df
```

```python
danceability_genre=spotify_df.groupby("track_genre")["danceability"].mean()
danceability_genre
```

```python
instrumental_genres=spotify_df[spotify_df["instrumentalness"]>0.8]
instrumental_genres[["artists","track_name","track_genre","instrumentalness"]]
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=spotify_df,x="explicit",y="popularity")
plt.show()
```

```python
spotify_df.info()
```

```python
numerical_features=["popularity","duration_min","danceability","energy","loudness","speechiness","acousticness","instrumentalness","liveness","valence","tempo"]
corr_features=spotify_df[numerical_features].corr()
```

```python
corr_features
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr_features,cmap="coolwarm",annot=True)
plt.show()
```
