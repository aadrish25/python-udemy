```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_data=pd.read_csv("covid-19.csv")
covid_data
```

```python
covid_data.loc[covid_data["Confirmed"].idxmax(),["Country/Region","Confirmed"]]
```

```python
avg_deaths=covid_data["Deaths"].mean()
print(f"Average deaths across all countries: {int(avg_deaths)}")
```

```python
recovery_rate=covid_data[["Country/Region","Recovered / 100 Cases"]].sort_values(by="Recovered / 100 Cases",ascending=False)
recovery_rate.head(5)
```

```python
covid_data.loc[covid_data["Confirmed"].idxmax(),["WHO Region"]]
```

```python
active_cases=covid_data[["Country/Region","Active"]].sort_values(by="Active",ascending=False)
active_cases.head(5)
```

```python
covid_data
```

```python
highest_weekly_percentage=covid_data.loc[covid_data["1 week % increase"].idxmax(),["Country/Region"]]
print(highest_weekly_percentage)
```

```python
countries_high_new_cases=covid_data[covid_data["New cases"]>1000]
countries_high_new_cases[["Country/Region","New cases"]]
```

```python
death_rate=covid_data[["Country/Region","Deaths / 100 Cases"]]
death_rate
```

```python
sns.histplot(data=death_rate,x="Deaths / 100 Cases")
plt.show()
```

```python
features=["Confirmed","Deaths","Recovered"]
correl=covid_data[features].corr()
correl
```

```python
sns.scatterplot(data=covid_data,x="Confirmed",y="Recovered")
plt.show()
```

```python
countries_no_recovery=covid_data.loc[covid_data["New recovered"]==0,["Country/Region","New recovered"]]
countries_no_recovery
```

```python
avg_active_by_region=covid_data.groupby("WHO Region")["Active"].mean()
avg_active_by_region
```

```python
countries_high_weeklyrise=covid_data.loc[covid_data["1 week % increase"]>20,["Country/Region","1 week % increase"]]
countries_high_weeklyrise
```

```python
weekly_increase_regionwise=covid_data.groupby("WHO Region")["1 week % increase"].mean().reset_index(name="Weekly increase %")
weekly_increase_regionwise
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=weekly_increase_regionwise,x="WHO Region",y="Weekly increase %")
plt.show()
```

```python
covid_data["Fatality Rate"]=(covid_data["Deaths"]/covid_data["Confirmed"])*100
covid_data
```

```python
highest_fatality_rates=covid_data[["Country/Region","WHO Region","Fatality Rate"]].sort_values(by="Fatality Rate",ascending=False)
highest_fatality_rates
```
